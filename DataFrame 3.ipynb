## Filtering with the where Method

```python
df = pd.read_csv('jamesbond.csv', index_col='Film')
df.sort_index(inplace = True)
df.head(2)
```

```python
mask = df['Actor'] == 'Sean Connery'
df[mask]
```

```python
df.where(mask)
```

```python
df.where(df['Budget'] > 100)
```

## The .query() Method

```python
df = pd.read_csv('jamesbond.csv', index_col='Film')
df.sort_index(inplace = True)
df.head(2)
```

```python
df.columns
```

```python
df.columns = [column_name.replace(' ', '_') for column_name in df.columns]
df.head(1)
```

```python
df.query(' Actor == "Sean Connery" ')
```
